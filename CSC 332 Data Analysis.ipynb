# Looking at Total Time Required, Average Waiting Time, and CPU Efficiency

```python
efficiency
```

```python
fig = plt.figure(figsize=[10,15])
ax1 = fig.add_subplot(3, 1, 1)
ax1.bar(x=efficiency.index, height=efficiency['Total Time Required'],color='red')
ax1.set_xlabel('Algorithm')
ax1.set_ylabel('Time in ms')
ax1.set_title('Comparing Total Time Required')
ax2 = fig.add_subplot(3, 1, 2)
ax2.bar(x=efficiency.index, height=efficiency['Average Waiting Time'], color='blue')
ax2.set_xlabel('Algorithm')
ax2.set_ylabel('Time in ms')
ax2.set_title('Comparing Average Waiting Time')
ax3 = fig.add_subplot(3, 1, 3)
ax3.bar(x=efficiency.index, height=efficiency['CPU Efficiency'], color='navy')
ax3.set_xlabel('Algorithm')
ax3.set_ylabel('Efficiency %')
ax3.set_title('Comparing CPU Efficiency')
plt.subplots_adjust( hspace=.3)
```

As we can see from the plots above the most notable observation is that the most efficient alogorthms in terms of CPU efficiency are FCFS and RR with a time Quantum of 100, with FCFS being just slighty more efficient than RR 100.

When looking at average waiting times, again FCFS has the lowest average waiting time with RR 100 again taking second place and RR 50 being the runner up. 

We see a similar trend with the total time required and see that overall FCFS is the best algorithm out of all of our CPU scheduling algorithms.
